```python
import os
from data_augmentation import draw_annotations, crop_images_and_annotations, augment_data
dataset_dir = "/Users/emirulurak/Desktop/dev/Inovako_folders/dataset_tofas/"
splits = ["train", "valid", "test"]
augment_dataset_dir = "/Users/emirulurak/Desktop/dev/Inovako_folders/augmented_dataset_tofas/"
```

```python
total_images = 0
for split in splits:
    print(len(os.listdir(f"{dataset_dir}{split}/images")))
    total_images += len(os.listdir(f"{dataset_dir}{split}/images"))
print(f"Total images: {total_images}")
```

```python
# # visualization of tofas dataset
for split in splits:
    for dir in os.listdir(f"{dataset_dir}{split}/images/"):
        img_dir = os.path.join(f"{dataset_dir}{split}/images/", dir)
        annot_dir = f"{dataset_dir}{split}/labels/{dir[:-3]}txt"
        img_aug_dir = os.path.join(f"{augment_dataset_dir}{split}/images/", f"{dir[:-4]}_aug_0.jpg")
        annot_aug_dir = os.path.join(f"{augment_dataset_dir}{split}/labels/", f"{dir[:-4]}_aug_0.txt")
        draw_annotations(img_dir=img_dir, annot_dir=annot_dir, aug=False)
        draw_annotations(img_dir=img_aug_dir, annot_dir=annot_aug_dir, aug=False)
```
